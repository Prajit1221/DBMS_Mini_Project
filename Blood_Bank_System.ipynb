Q1. Recipients and Donors names having the same blood group registered on the same date.

```python
%%sql
select Blood_Donor.bd_name,Recipient.reci_name,reco_Name from Recording_Staff  
inner join Blood_Donor on Recording_Staff.reco_ID = Blood_Donor.reco_ID 
inner join Recipient on Recording_Staff.reco_ID = Recipient.reco_ID where Blood_Donor.bd_Bgroup = Recipient.reci_Brgp
and Blood_Donor.bd_reg_date = Recipient.reci_reg_date
```

Q2. Show the blood specimen verified by disease finder Mark which are pure (status=1).

```python
%%sql
Select specimen_number,b_group from BloodSpecimen,DiseaseFinder 
WHERE BloodSpecimen.dfind_ID= DiseaseFinder.dfind_ID AND dfind_name='Mark' AND status=1;
```

Q3. List all recipients with their blood group, quantity, and registration date.

```python
%%sql
SELECT reci_name, reci_Brgp, reci_Bqnty, reci_reg_date
FROM Recipient;
```

Q4. Show the donors having the same blood groups required by the recipient staying in the same city along with recipient details.

```python
%%sql
Select bd_ID,bd_name,reci_ID,reci_name FROM Blood_Donor,Recipient
WHERE bd_Bgroup=reci_Brgp AND Blood_Donor.City_ID= Recipient.City_ID
```

Q5.Display the information of Hospital_Info_1 handled by BB_Manager whose ID is 103

```python
%%sql
Select hosp_ID,hosp_name , City_ID, Hospital_Info_1.M_id from Hospital_Info_1,BB_Manager 
where BB_Manager.M_id=Hospital_Info_1.M_id and BB_Manager.M_id=103;
```

Q6. Find the number of blood specimens available for each blood group.

```python
%%sql
select sum (specimen_count) from (SELECT b_group, COUNT(*) AS specimen_count
FROM BloodSpecimen
WHERE status = 1
GROUP BY b_group);
```

Q7. List the hospitals along with their needed blood groups and quantities.

```python
%%sql
SELECT h.hosp_name, h2.hosp_needed_Bgrp, h2.hosp_needed_qnty
FROM Hospital_Info_1 h
JOIN Hospital_Info_2 h2 ON h.hosp_ID = h2.hosp_ID;
```

Q8. Find the blood donors who have donated blood to a recipient with blood group AB+.

```python
%%sql
SELECT bd_name
FROM Blood_Donor bd
JOIN BloodSpecimen bs ON bd.bd_Bgroup = bs.b_group
WHERE bs.dfind_ID IN (
    SELECT BloodSpecimen.dfind_ID
    FROM BloodSpecimen
    JOIN Recipient ON BloodSpecimen.M_id = Recipient.M_id
    WHERE Recipient.reci_Brgp = 'AB+'
);
```

Q9. List the recipients and donors who are in the same city.

```python
%%sql
SELECT r.reci_ID, r.reci_name, r.reci_Brgp, bd.bd_name,c.city_name
FROM Recipient r
JOIN City c ON r.City_ID = c.City_ID
JOIN Blood_Donor bd ON c.City_ID = bd.City_ID
WHERE r.reci_Bqnty > 0;
```
